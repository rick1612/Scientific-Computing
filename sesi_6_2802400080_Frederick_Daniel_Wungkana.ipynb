```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')
%matplotlib inline

#define parameters
f = lambda t, s: np.exp(-t)
h = 0.1
t = np.arange(0, 1 + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Persamaan diferensial: y' = -4y + x^2
f = lambda x, y: -4*y + x**2

# Langkah dan domain
h = 0.1
x = np.arange(0, 1 + h, h)
y0 = 1

# Metode Euler
y = np.zeros(len(x))
y[0] = y0

for i in range(len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

# Solusi eksak
# y(x) = (1/4)x^2 - (1/8)x + 1/64 + (63/64)e^(-4x)
exact = lambda x: (1/4)*x**2 - (1/8)*x + (1/64) + (63/64)*np.exp(-4*x)

# Plot
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo--', label='Approximate (Euler)')
plt.plot(x, exact(x), 'g', label='Exact Solution')
plt.title('Approximate and Exact Solution for y\' + 4y = x^2')
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.legend(loc='best')
plt.show()
```
